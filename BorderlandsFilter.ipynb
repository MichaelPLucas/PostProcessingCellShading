## Getting Started

   So to get started I'm going to need some imports and I'm also going to need to grab the image and some information about it. That's all pretty simple to do and doesn't need a whole lot of explanation so I'll just put the lines of code necessary under this block and move on.
```python
%matplotlib inline
import matplotlib.pyplot as plt
from skimage import filters
from skimage import feature
from skimage import color
from scipy import ndimage
import numpy as np

start = plt.imread("lenna.png")
pixels = len(start) * len(start[0])

plt.imshow(start)
plt.show()
```

## The Method

### Step One

   So we have some important imports and our initial image. I've also provided a few starter images for testing within the same directory as this notebook as a courtesy to anyone who wants to try this filter out. We also calculated the number of pixels in the image by multiplying the height by the width and stored that for future use.
   
   So how do we make something look like it's from the Borderlands games, or even just look cell shaded generally? One of the defining characteristics that immediately came to mind for me is that cell shaders tend to have flat dominant colors without any lambert shading. I figured a good way to convert an image to look more like that would be to use image segmentation to pull out the dominant colors in the image. This will also help in creating the heavy outlines characteristic of the Borderlands art style.
   
   To do this, I've written my own implementation of the K Means algorithm. The specifics of how the algorithm works are not terribly important for this, but essentially it finds the dominant colors and segments the image through distance based grouping. We can use this to segment our image, which is step one in getting making our Borderlands filter.

```python
def avg_pxl(arr):
    r = sum([p[0] for p in arr])
    g = sum([p[1] for p in arr])
    b = sum([p[2] for p in arr])
    return [r / len(arr), g / len(arr), b / len(arr)]

def dist(p1, p2):
    r = abs(p2[0] - p1[0])
    g = abs(p2[1] - p1[1])
    b = abs(p2[2] - p1[2])
    return r + g + b

def k_means(pxls, centers):
    result = {}
    
    for x in range(len(pxls)):
        for y in range(len(pxls[x])):
            best = 0
            bdist = -1
            for i in range(len(centers)):
                ndist = dist(centers[i], pxls[x][y])
                if (bdist == -1 or ndist < bdist):
                    best = i
                    bdist = ndist
            result[(x, y)] = best
    
    return result

centers = []
for i in range(3):
    x = len(start) * (2 * i + 1) / 6
    for j in range(3):
        y = len(start[0]) * (2 * j + 1) / 6
        centers.append(start[x][y])

last = []
clusters = {}
while not last == centers:
    last = centers
    clusters = k_means(start, centers)
    
    temp = []
    for i in range(16):
        temp.append([])
    
    for key, value in clusters.items():
        temp[value].append(start[key[0]][key[1]])
    
    for i in range(len(temp)):
        if (len(temp[i]) == 0):
            continue
        centers[i] = avg_pxl(temp[i])

for x in range(len(start)):
    for y in range(len(start[x])):
        start[x][y] = centers[clusters[(x, y)]]

plt.imshow(start)
plt.show()
```

### Step Two

   We now have a segmented image which has very visibly been reduced to only the dominant colors from the image. It is worth noting that you can choose how many segments are created by the K Means algorithm, and that number is actually the 'K' in 'K Means'. For this application I chose 9 segments with initial centers on the intersections of the quarter lines of the image.
   
   So now that we have that done, we'll need to start working on another interesting facet of the Borderlands art style: the heavy black outlines around just about everything. Gearbox - the developers of Borderlands - revealed a while back that this part of their art style is actually done as a post processing effect, so we have a chance at getting really close to their exact style with this one. My best guess for how to handle this aspect of the art style is to find our edges with a sobel filter and then black out the edges manually. To do this, we're going to need to convert to grayscale to run the sobel filter, and we're also going to need to blur the image slightly to remove some of the noise generated by segmenting the image. All of this noise would show up as edges, which would give us a lot of random looking black marks. This also helps in making the edges thicker and giving a heavier outline to the image.

```python
blurred = filters.gaussian(start, pixels / 400000.0, multichannel=True)

gray = color.rgb2grey(blurred)

dx = ndimage.sobel(gray, 0)  # horizontal derivative
dy = ndimage.sobel(gray, 1)  # vertical derivative
mag = np.hypot(dx, dy)  # magnitude
mag *= 255.0 / np.max(mag)  # normalize (Q&D)
mag = ndimage.filters.convolve(mag, [[3/18.0, 1/18.0, 3/18.0], [1/18.0, 2/18.0, 1/18.0], [3/18.0, 1/18.0, 3/18.0]])

plt.imshow(mag)
plt.show()
```

### Step Three
    
   As you can see we have our edges now. Combining those with our original image we can get pretty close to our final product. We're going to want to apply one last blur to the image to remove some of the noise from our edge finding algorithm and clean up the finished product slightly, but that's really all it took to go from run of the mill image to Borderlands style art. The last if statement is a correction in the event that the image is in grayscale at the end of the filtering process, which would be important for outputting the image to a file. So there we have it! A three step process for filtering something to look like it was cell shaded. I unfortunately couldn't find a good scaling for the last blur that looked good on all resolutions of images, but overall I'm pretty happy with how well it works on both high and low res images.

```python
end = start
for i in range(len(end)):
    for j in range(len(end[0])):
        if (mag[i][j] > 45):
            if (type(end[i][j]) is np.array):
                end[i][j] = [0, 0, 0]
            else:
                end[i][j] = 0

end = filters.gaussian(end, pixels / 700000.0, multichannel=True)

if (type(end[0][0]) is not np.ndarray):
    end = [[[x, x, x] for x in row] for row in end]

plt.imshow(end)
plt.show()
```
